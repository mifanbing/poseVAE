```python
import tensorflow as tf

from tensorflow.keras.layers import Input, Dense, Reshape, Flatten, Dropout, Layer
from tensorflow.keras.layers import BatchNormalization, Activation, Embedding, multiply, Concatenate
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import UpSampling2D, Conv2D
from tensorflow.keras.layers import Conv2DTranspose

from tensorflow.keras import activations


from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras import backend
from tensorflow.keras.initializers import RandomNormal
from tensorflow.keras.constraints import Constraint
from tensorflow.keras.optimizers import RMSprop, Adam

import numpy as np
# import tensorflow as tf
import math
from matplotlib import pyplot
import os

class Sampling(Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

def reLoss(data, reconstruction):
      #print(data.shape)
      reconstruction_loss = tf.reduce_mean(
          tf.reduce_sum(
              tf.keras.metrics.mean_squared_error(data, reconstruction), axis=(1, 2)
          )
      )

      tf.print(reconstruction_loss)

      return reconstruction_loss 

class VAE():
    def __init__(self):
        # Input shape
        self.img_rows = 96
        self.img_cols = 96
        self.channels = 3
        self.img_shape = (self.img_rows, self.img_cols, self.channels)
        self.latent_dim = 400

        self.img_names = []
        pathes = ["./drive/MyDrive/PoseProject/Data/imageyeFilter", 
                  "./drive/MyDrive/PoseProject/Data/imageye2Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye3Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye4Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye5Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye6Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye7Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye8Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye9Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye10Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye11Filter",
                  "./drive/MyDrive/PoseProject/Data/imageye12Filter"
                ]
        for path in pathes:
            for imageName in os.listdir(path):
              path2 = os.path.join(path, imageName)
              self.img_names.append(path2)

        lr_schedule = tf.keras.optimizers.schedules.PolynomialDecay(
            0.001,
            decay_steps = 10000,
            end_learning_rate = 0.0001,
            power = 3.0)
        opt = Adam(learning_rate=lr_schedule)
        #opt = Adam(learning_rate=0.0001)

        self.decoder = self.build_decoder()
        self.encoder = self.build_encoder()

        inputImage = Input(shape = self.img_shape)
        z_mean, z_log_var, latentVector = self.encoder(inputImage)
        kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
        kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))

        reconstructImage = self.decoder(latentVector)
        self.combined = Model(inputImage, reconstructImage)
        self.combined.add_loss(kl_loss)
        self.combined.compile(loss = reLoss, optimizer = opt)

    def build_decoder(self):
        noise = Input(shape=(self.latent_dim,))
        noise_conv = Dense(3 * 3 * 650, activation="relu", input_dim = self.latent_dim)(noise)
        noise_conv = Reshape((3, 3, 650))(noise_conv)

        gen = Conv2DTranspose(512, (3,3), strides=(2,2), padding='same')(noise_conv)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(256, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(128, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)  
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(64, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)  
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(32, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)  
        gen = BatchNormalization()(gen, training = True)

        relu_custom = lambda x: activations.relu(x, max_value=1)

        # output
        out_layer = Conv2D(3, (1,1), activation=relu_custom, padding='same')(gen)
 
        generator = Model(noise, out_layer)    

        return generator

    def build_encoder(self):
        img = Input(shape = self.img_shape)

        fe = Conv2D(32, (3,3), strides=(2,2), padding='same')(img)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)
        #fe = Dropout(0.2)(fe)

        fe = Conv2D(64, (3,3), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)
        #fe = Dropout(0.2)(fe)

        fe = Conv2D(128, (3,3), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)
        #fe = Dropout(0.2)(fe)

        fe = Conv2D(256, (3,3), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)
        #fe = Dropout(0.2)(fe)

        fe = Conv2D(512, (3,3), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)
        #fe = Dropout(0.2)(fe)

        fe = Flatten()(fe)
        #fe = Dropout(0.2)(fe)

        z_mean = Dense(self.latent_dim)(fe)
        z_log_var = Dense(self.latent_dim)(fe)
        z = Sampling()([z_mean, z_log_var])

        model = Model(img, [z_mean, z_log_var, z])

        return model

    def load_batch_imgs(self,batch_size,shuffle_ind,round):
        img = []
        
        for i in range(batch_size):
            index = shuffle_ind[round*batch_size+i]
            images = image.load_img(self.img_names[index], target_size=(self.img_rows,self.img_cols))
            x = image.img_to_array(images)
            x = np.expand_dims(x, axis=0)
            img.append(x)

        x = np.concatenate([x for x in img])
        x = x / 127.5 - 1.

        return x

    def train(self, epochs, batch_size, save_interval, log_interval):
        sample_num = len(self.img_names)
        bat_per_epo=math.floor(sample_num/batch_size)
        lossHistory = []

        for epoch in range(epochs):   
            for j in range(bat_per_epo):
              shuffle_ind = np.random.permutation(sample_num)
              imgs = self.load_batch_imgs(batch_size, shuffle_ind, j)
              
              loss = self.combined.train_on_batch(imgs, imgs)
              print("%d %d total loss: %f" % (epoch, j, loss))
            
              if epoch % log_interval == 0 and j == 0:
                  lossHistory.append(loss)
                  pyplot.plot(lossHistory, label = "loss")
                  pyplot.ylim([0, 5000])
                  pyplot.show()

              if epoch % save_interval == 0 and j == 0:
                  vectors = self.encoder(imgs)
                  vector0 = tf.reshape(vectors[2], [64, self.latent_dim]).numpy()
                  print("var %f" % (np.var(vector0)))
                  print("mean %f" % (np.mean(vector0)))
                  # for vector in vector0:
                  #   print(vector)

                  # _ = pyplot.hist(vector0, bins='auto')
                  # pyplot.show()

                  recontructImages = self.combined.predict(imgs)
                  self.save_imgs(recontructImages, epoch)
                  self.saveModel()

    def save_imgs(self, recontructImages, epoch):
        r, c = 8, 8
        
        fig, axs = pyplot.subplots(r, c)
        cnt = 0
        for i in range(r):
            for j in range(c):
                axs[i, j].imshow(recontructImages[cnt])
                axs[i, j].axis('off')
                cnt += 1

        fig.savefig("./drive/MyDrive/PoseProject/result/result_%d.png" % epoch)
        pyplot.close()

        # cnt = 0
        # for i in range(r):
        #     for j in range(c):
        #         pyplot.imshow(recontructImages[cnt])
        #         pyplot.savefig('./drive/MyDrive/PoseProject/result/aa_%d.png' %(cnt))
        #         cnt += 1

        # pyplot.close()

    def loadModel(self):
        self.decoder.load_weights('./drive/MyDrive/PoseProject/result/decoderWeight.h5')
        self.encoder.load_weights('./drive/MyDrive/PoseProject/result/encoderWeight.h5')
        self.combined.load_weights('./drive/MyDrive/PoseProject/result/combinedWeight.h5')

    def saveModel(self):
        self.combined.save_weights("./drive/MyDrive/PoseProject/result/combinedWeight.h5")
        self.encoder.save_weights("./drive/MyDrive/PoseProject/result/encoderWeight.h5")
        self.decoder.save_weights("./drive/MyDrive/PoseProject/result/decoderWeight.h5")

vae = VAE()
#vae.loadModel()
vae.train(epochs=2000, batch_size=64, save_interval = 10, log_interval = 5)

```
